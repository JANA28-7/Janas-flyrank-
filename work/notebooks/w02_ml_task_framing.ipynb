## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Lane: CTR / Engagement Opportunity Scoring

ML task type: Ranking / Scoring.

The goal is to rank pages based on their opportunity for CTR improvement. Instead of predicting a simple yes/no outcome, the model assigns a score that helps prioritize which pages should be reviewed first.


## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target is the CTR (Click Through Rate) or a score representing CTR improvement opportunity.
 Pages with high impressions but relatively low CTR are considered higher priority for review.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The success of the model will be measured by how well it ranks the pages with the biggest opportunity for improvement. Precision@K is an appropriate metric because the goal is to prioritize the best pages for review.

## 5. Why ML beats a fixed rule here
